<h3>Manipulación de dataframes <h3/>

```python
# Cambiar la columna date time para separarla en hora y fecha
planta1['DATE_TIME']=pd.to_datetime(planta1['DATE_TIME'], format='%d-%m-%Y %H:%M')
planta1=planta1.groupby('DATE_TIME').sum().reset_index()
planta1["TIME"]=planta1["DATE_TIME"].dt.time
planta1["DATE"]=planta1["DATE_TIME"].dt.date
```

```python
# Eliminar la columna que ya no se usa
planta1.drop("DATE_TIME", axis=1, inplace=True)
```

```python
planta1
```

```python
link2="https://raw.githubusercontent.com/UCPECO/Algoritmos_UCP/master/pandas_matplotlib_CERDS/data/Plant_1_Weather_Sensor_Data.csv"
sens=pd.read_csv(link2)
```

```python
# Mismas operaciones que el dataframe anterior
sens['DATE_TIME']=pd.to_datetime(sens['DATE_TIME'], format='%Y-%m-%d %H:%M:%S')
sens=sens.groupby('DATE_TIME').sum().reset_index()
sens["TIME"]=sens["DATE_TIME"].dt.time
sens["DATE"]=sens["DATE_TIME"].dt.date
```

```python
sens.drop("DATE_TIME", axis=1, inplace=True)
```

```python
sens
```

<h3>Creación de gráficos<h3/>

```python
# Fijar un estilo de gráficos, existen más estilos
plt.style.use("ggplot")
```

```python
# Creación del contener para poder hacer una gráfica
fig, ax=plt.subplots(dpi=100, figsize=(14,8))
#Fijar títulos de cada uno de los ejes y el título general
plt.ylabel("Rendimiento diario")
plt.xlabel("Fecha")
plt.title("Rendimiento diario por fecha")
# Gráfica de barras
ax.bar(x=planta1["DATE"], height=planta1['DAILY_YIELD'], width=0.4)
```

```python
fig, ax=plt.subplots(dpi=100, figsize=(14,8))
# Gráficas sobre una misma figura
planta1.plot(ax=ax, x="TIME", y="AC_POWER", style="o")
planta1.plot(ax=ax, x="TIME", y="DC_POWER", style="o")
plt.ylabel("DC y AC")
plt.xlabel("Hora del día")
plt.title("Comparación entre DC y C")
#Crear leyenda del gráfico
plt.legend()
```

```python
# Definición de un contenedor con dos columnas
fig, ax=plt.subplots(ncols=2, nrows=1,dpi=100, figsize=(20,6))
sens.plot(x='TIME', y='IRRADIATION', ax=ax[1], style='d')
# Gráfica obtenido el promedio de una columna
sens.groupby("TIME")['MODULE_TEMPERATURE'].agg('mean').plot(ax=ax[0])
sens.groupby("TIME")['AMBIENT_TEMPERATURE'].agg('mean').plot(ax=ax[0])
```

```python
# Seaborn permite hacer un poco más que matplotlib
import seaborn as sns
```

```python
# Crear un dataframe que combine los dos anteriores
merged=sens.merge(planta1, on=["TIME", "DATE"])
```

```python
f=plt.figure(dpi=100)
# Crear un mapa de correlación exluyendo las columnas de fecha y tiempo
sns.heatmap(merged.drop(columns=['DATE', 'TIME']).corr(), annot=True)
plt.show()
salvar la figura en archivo tipo png
f.savefig("correlacion.png")
```
