<h3><center>¿Qué es Pandas y Matplotlib?<center/><h3/>

<p align="justify"> Pandas y Matplotlib son dos librerías alojadas en Python. Su uso complementario permite realizar gráficos para una gran cantidad de información, sobre todo a partir del surgimiento de la era digital y con esta, la masificación de los datos.<p/> 
<p align="justify">Los datos pueden estar contenidos en distintos tipos de formatos, pero uno de los formatos más famosos son los archivos "csv" o "valores separados por coma", permitiendo almacenar los datos en columnas y con espacio de almacenamiento menor que otro tipo de arhivos. <p/>
<p align="justify"> Para conocer las funciones básicas de Pandas y Matplotlib, se usará un conjunto de datos que describen la energía producida por dos plantas de energía solar durante un periodo de 34 días.
